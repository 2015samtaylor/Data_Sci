### Requirement FR1.1 - Find the arithmetic mean

```python
# Define two variables, one that creates the sum, and the other which takes the count of the list. Then divide the two.
# lastly round the number to the nearest whole number for cleanliness. 

def mean_eval(data):
    total = sum(data)
    length = len(data)
    mean = int(round(total/length, 1))
    return(mean)

# must initialize as a global variable so I can use in later steps. 
mean = mean_eval(nums)
length = len(nums)

print("The mean value is: " + str(mean))

# global function 
mean_eval(nums)
 
```

### Requirement FR1.2 - Find the standard deviation

```python
# step one - subtract each point of data from the original mean
# step two - once the result is found square each number in the set. 
# step three - after squaring all the numbers find the mean again by utilizing a while loop.
# step four - finally find the square root of the new mean, and round to the nearest tenth

def std_eval(data):
    step_one = [dt-mean for dt in data]
    step_two = [dt**2 for dt in step_one]
    index = 0
    length = len(step_two)

    total = 0
    while length > index:
        total += step_two[index]
        index += 1

    new_mean = (total/(length))
    
    standard_deviation = (new_mean**.50)
    standard_deviation = int(round(standard_deviation, 1))
    
    return(standard_deviation)

# establish standard deviation as a global variable to be mentioned in the summary
standard_deviation = std_eval(nums)
sd = mean - standard_deviation
sdd = mean + standard_deviation
    
print('Assuming the data is normally distributed, about 2/3 of the data fall within ' + str(sd)+ ' and ' + str(sdd))
print("The standard deviation is: " + str(standard_deviation))

# global function 
std_eval(nums)
```

<p style="color:red; font-weight:bold">Requirement is met to a professional standard, producing accurate results. Nice touch with the extra printed information and good pseudocode comments. </p>

### Requirement FR1.12 - Create and demonstrate a function to print a custom table

```python
# created a function which specifies a given length when called upon. If the length is short spaces are padded,
#  and if the length is too long the text is cut down. 

def fixed_length(string, length):
    if len(string) > length:
        string = string[:length]
    elif len(string) < length:
        string = (string + ' ' * length)[:length]
    return string

# must change integers to strings in order to be subbed into the function below,


# the names of the columns, and functions of the values variable are iterated through in a for loop within the 
# function stat_summ
column = ('Count', 'Mean', 'STD', 'Min', '25%', '50%', '75%', 'Max')

values = (length, mean, standard_deviation, minimum, ind_25, ind_50, ind_75, maximum)

str_values = list(map(str, values))

# a function within a function in order to get a stat summary.

def stat_summ():
    for words in column:
        print(fixed_length(words, 8), end = ' ')
    print()
    for nums in str_values:
        print(fixed_length(nums, 8), end = ' ')
        
(stat_summ())
```

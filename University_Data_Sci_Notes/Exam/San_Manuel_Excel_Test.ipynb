# Part 3: Transformation

```python
# Add a field for unit price. Unit price = Sales_Amount / Quantity.

df['Unit_Price'] = df['Sales_Amount'] / df['Quantity']

df
```

```python
# Which SKU has the highest unit price?

# SKU with the highest unit price is E3T4K. 

df.sort_values(by=['Unit_Price'], ascending = False)
```

```python
# This SKU was sold at another price on another occasion. What was the other unit price?

# This SKU was sold at two different times. Once for 277.62, and once for 322.23

df['SKU'] = df['SKU'].astype(str)


SKU = df.loc[df['SKU'] == 'E3T4K']

SKU
```

```python
# Add a field called "Big Sale" that will have a value of 1 if the unit price is greater than the average 
# unit price and with a quantity greater than 1, otherwise set the value to 0.


average = df['Unit_Price'].mean()

# create big sales category with lambda function. 
df['Big Sales'] = df['Unit_Price'].apply(lambda x: 1 if x > average else 0)

df
```

```python
# How many records have a Big Sale?

print(df['Big Sales'].value_counts())


print("There are 4832 records that have a big sale.")
```

```python
# Create a new sheet called "Daily Sales". On this new sheet, we will create a summarized dataset 
# of daily sales record with one record per date for each day in 2016. 
# This will be 366 rows since there are 366 days in 2016. 
# This dataset will have the following fields for now: Date, Quantity Sold, Total Sales. 
# Quantity Sold is the sum of Quantity, and Total Sales is the sum of Sales_Amount for the given date. 
# Hint: Use the SUMIF function.
```

```python
df
```

```python
Daily_Sales = df[['Date', 'Quantity', 'Sales_Amount']]
```

```python
Ind_Days = Daily_Sales.groupby(pd.Grouper(key='Date', axis = 0, freq = 'D')).sum()
```

```python
# What are the top 5 dates for Total Sales in descending order? Include the total sales for each date.

Ind_Days.sort_values(by='Sales_Amount', ascending = False).head(5)
```

```python
# What is the average total sales amount per day?

Ind_Days['Sales_Amount'].mean()

print("The average sales amount per day is $619.82")
```

```python
# What is the median total sales amount per day?

Ind_Days['Sales_Amount'].median()

print("The median sales amount is 616.05")
```

```python
# What is the maximum quantity sold in a day?

Ind_Days['Quantity'].max()

print("The maximum quantity sold in one day was 473.")
```

```python
# Add a column indicating day of the week to your daily sales dataset (Monday, Tuesday, etc.). 
# Create a pivot table from the daily sales dataset. Place it in a new sheet called "DOW Pivot". 
# In this pivot table, we want to see the average total sales and average quantity sold by day of week.
```

```python
Ind_Days = Ind_Days.reset_index()
```

```python
Ind_Days['Day_of_Week'] = Ind_Days['Date'].dt.day_name()
```

```python
# New DataFrame with Day_of_Week

Ind_Days
```

```python
# In this pivot table, we want to see the average total sales and average quantity sold by day of week.

Dow_Pivot = Ind_Days.groupby(Ind_Days['Day_of_Week']).mean().reset_index()


Dow_Pivot
```

```python
# Which day of the week has the lowest average total sales?

Dow_Pivot.sort_values(by = 'Sales_Amount')

print("The day of the week with the lowest total sales in Sunday at 397.33")
```
